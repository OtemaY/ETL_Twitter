## Twitter Scraping

Writing a script that downloads tweets data on a specific search topic using the standard search 
Chosen topic: December21st

```python
#Importing libraries

import numpy as np
import pandas as pd
import os
import tweepy
import credentials as c
from tweepy import OAuthHandler
from datetime import datetime, time
```

```python
#Authorization


auth = tweepy.OAuthHandler(c.api_key, c.api_secret_key)
auth.set_access_token(c.access_token  ,c.access_token_secret)
api = tweepy.API(auth, wait_on_rate_limit=True)

   
text_query = 'December21st'

# Writing scrapetweets() function


#Scrape_tweets():
    
tweets = tweepy.Cursor(api.search, q=text_query, lang='en').items(19)

def scrape_tweets():
    Funnytweets = []
    for tweet in tweets:
        Funnytweets.append(
            {"Tweet": tweet.text, 
             "id":tweet.id,
             "source":tweet.source,
             "coordinates":tweet.coordinates,
             "retweetcount": tweet.retweet_count,
             "likescount" : tweet.favorite_count,
             "username": tweet.user.name,
             "screenname":tweet.user.screen_name,
             "friendscount":tweet.user.friends_count,
             "verified":tweet.user.verified,
             "description":tweet.user.description,
             "followerscount":tweet.user.followers_count
            })
    FunnyTweets = pd.DataFrame(Funnytweets)
    return FunnyTweets

FunnyTweets = scrape_tweets()
FunnyTweets
```
